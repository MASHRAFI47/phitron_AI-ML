### Variable

```python
number = 15151
height = 4.6
name = "Akash"
is_passed = True

print(type(number), type(height), type(name), type(is_passed))
```

```python
name = input("What is your name?")

print(type(name), name)
```

```python
age = int(input("What is your age?"))

print(type(age), age)

age += 3
print(age)
```

```python
x = 10 ; y = 3

sum = x+y
sub = x-y
mult = x*y
div = round(x / y, 2)
rem = x % y

print(sum, sub, mult, div, rem)
```

```python
x = 10 ; y = 3

greater_than = x > y
greater_than_or_equal = x >= y

print(greater_than, greater_than_or_equal)

a = 3; b = 3

equal = a==b
not_equal = a != b
print(equal, not_equal)
```
