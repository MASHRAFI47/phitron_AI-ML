### For loop

```python
for i in range(10):
    print(i)
```

```python
for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(1, 10):
    for j in range(1, 5):
        print(i, j)
```

### While loop

```python
count = 1

while count < 10:
    print(count, "Hello world")
    count += 1

print(count)
```

```python
sum = 0

for i in range(1, 11):

    print(i, "is processing")

    if i%2 == 0:
        continue

    print(i, "is added in the sum")
    sum += i

print(sum)
```

```python
accuracy = 95

for i in range(20):
    accuracy += 1;
    print(accuracy)

    if accuracy == 100:
        break;
```
